Description: give two sides of a right triangle, print the other side`

```python
from lzxPyLib.StrongerInput import *
from lzxPyLib.NumRangeCheck import *

import math

option = 0.0
isExit = False

while isExit == False:
    print("+------------------------+")
    print("| Pythagoras' Calculator |")
    print("| 1 - Find hypotenuse    |")
    print("| 2 - Find another side  |")
    print("| 9 - Exit               |")
    print("+------------------------+")
    
    option = IntInput()
    
    #get out of the program if the user inputs 9
    if option == 9:
        print("GoodBye!")
        isExit = True
        
    elif option == 1:
        while True:
            sideA = FloatInput("Input the first side : ")
            if(isNegetive(sideA)):
                print("triangle's side can't be negetive!")
                continue

            sideB = FloatInput("Input the second side: ")
            if(isNegetive(sideB)):
                print("triangle's side can't be negetive!")
                continue
            
            break
                
        hypotenuse = math.sqrt(sideA ** 2 + sideB ** 2)

        print("the hypotenuse is " + str(hypotenuse))
    
    elif option == 2:
        while True:
            sideA = FloatInput("Input the first side: ")
            if(isNegetive(sideA)):
                print("triangle's side can't be negetive!")
                continue
                
            hypotenuse = FloatInput("Input the hypotenuse: ")
            if(isNegetive(hypotenuse)):
                print("triangle's side can't be negetive!")
                continue
                
            if(compare(sideA, hypotenuse) == 1):
                print("hypotenuse should be longer than either short side")
                continue
                
            break

        sideB = math.sqrt(hypotenuse ** 2 - sideA ** 2)
        
        print("the othe short side is " + str(sideB))
        
    else:
        print("Please enter one of the give numbers")
        pass
```

Description: reverse an array<br>

```python
import numpy as np
temp = 0

# array = [1,4,2,5,7,3,7,9,8,10]
array=[]
for i in range (1,10):
    array.append(np.random.randint(1,10000))

quit = False
for i in range (1, len(array)):
    quit = True
    for j in range (0, len(array) - i):
        if array[j] > array[j + 1]:
            array[j], array[j + 1] = array[j + 1], array[j]
            quit = False
    if quit:
        break
print(array)

arrayClone = array
for i in range (0, int(len(arrayClone) / 2)):
    arrayClone[i], arrayClone[len(array) - i - 1] = arrayClone[len(arrayClone) - i - 1], arrayClone[i]
print(arrayClone)
```

Description: generate 2 different random numbers, apply 4 calculations +-*/ to them and print the answer

```python
import random

isFinished = False
while not isFinished:
    firstNum = random.randint(1, 100)
    secondNum = random.randint(1, 100)

    print(firstNum, "+", secondNum, "=", firstNum + secondNum)
    print(firstNum, "-", secondNum, "=", firstNum - secondNum)
    print(firstNum, "*", secondNum, "=", firstNum * secondNum)
    print(firstNum, "/", secondNum, "=", firstNum / secondNum)
    
    answer = input("Do you want to continue(y/n): ")
    if answer == "n":
        isFinished = True
    elif answer == "y":
        isFinished = False
    else:
        print("Invalid answer! continue as default")
```

Author: Michael 3778<br>
Date: 2017/9/25<br>
Description: print a menu and allow the user to convert bin to (hex, den, oct)

```python
from lzxPyLib.StrongerInput import *;

def BinToHex():
    #uses BinInputAsStr() function in lzxPyLib.StrongerInput
    binary = BinInputAsStr()
    
    #if the user enters nothing, give defaut value "0"
    if binary == "":
        binary = "0"
        
    #fill the missing digits with 0 if the binary entered don't have enough digits
    if len(binary) % 4 != 0:
        for i in range(0, 4 - (len(binary) % 4)):
            binary = "0" + binary
    
    #used to store the final answer
    hexadecimal = ""
    
    #used to store the current hex digit (4 binary digit -> 1 hex digit)
    currentHex = 0
    
    #used to store the current position iterated in the string
    positionPointer = 0
    
    #iterate through the binary string backwards with step 1
    for currentBit in binary[::-1]:
        
        if positionPointer <= 3:
            currentHex += int(currentBit) * (2 ** positionPointer)
            positionPointer += 1
        if positionPointer > 3:
            if currentHex <= 9:
                hexadecimal = str(currentHex) + hexadecimal
            else:
                #A -> 65 & currentHex of A is 10, so the char should be 55 + currentHex
                hexadecimal = chr(55 + currentHex) + hexadecimal
            positionPointer = 0
            currentHex = 0
    print("the hexadecimal form of", binary, "is", hexadecimal)

def BinToOct():
    #uses BinInputAsStr() function in lzxPyLib.StrongerInput
    binary = BinInputAsStr()
    
    #if the user enters nothing, give defaut value "0"
    if binary == "":
        binary = "0"
    
    #fill the missing digits with 0 if the binary entered don't have enough digits
    if len(binary) % 3 != 0:
        for i in range(0, 3 - (len(binary) % 3)):
            binary = "0" + binary
            
    #used to store the final answer
    octdecimal = ""
    currentOct = 0
    
    #used to store the current position iterated in the string
    positionPointer = 0
    
    #iterate through the binary string backwards with step 1
    for currentBit in binary[::-1]:
        if positionPointer <= 2:
            currentOct += int(currentBit) * (2 ** positionPointer)
            positionPointer += 1
        if positionPointer > 2:
            octdecimal = str(currentOct) + octdecimal
            positionPointer = 0
            currentOct = 0
            
    print("the octdecimal form of", binary, "is", octdecimal)
    
def BinToDen():
    #uses BinInputAsStr() function in lzxPyLib.StrongerInput
    binary = BinInputAsStr()
    
    #if the user enters nothing, give defaut value "0"
    if binary == "":
        binary = "0"
        
    #used to store the final answer
    denary = 0
    
    #used to store the current position iterated in the string
    positionPointer = 0
    
    #iterate through the binary string backwards with step 1
    for currentBit in binary:
        denary += int(currentBit) * (2 ** positionPointer)
        positionPointer += 1
    print("the denary form of", binary, "is", denary)
    
def PrintMenu():
    print("MAIN MENU")
    print("1. Binary to Hexadecimal")
    print("2. Binary to Ocatal")
    print("3. Binary to Denary")
    print("4. Exit")
    
def MainLoop():
    while True:
        PrintMenu()
        choice = IntInput("Enter your choice: ", "Please enter a integer!")
        if choice == 1:
            BinToHex()
        elif choice == 2:
            BinToOct()
        elif choice == 3:
            BinToDen()
        elif choice == 4:
            print("Thanks for using")
            break
        else:
            print("Not an option, enter again please")
            continue

        print()
        print("============================================")
        print("One operation finished, returning to menu...")
        print("============================================")
        print()

#=================[test field]================
MainLoop()
```

```python
def ValidatePassword(Pass=""):
    lowerCaseAmount = 0
    upperCaseAmount = 0
    numberAmount = 0
    for char in Pass:
        if ord(char) >= 65 and ord(char) <= 90:
            upperCaseAmount += 1
        elif ord(char) >= 97 and ord(char) <= 122:
            lowerCaseAmount += 1
        elif ord(char) >= 48 and ord(char) <= 51:
            numberAmount += 1
        else:
            return False
    if lowerCaseAmount < 2 or upperCaseAmount < 2 or numberAmount < 3:
        return False
    else:
        return True
    
testInputs = ["AAbb123", "Abb123", "AAbb12", "AAbb123#"]
for i in testInputs:
    print(ValidatePassword(i))
```
